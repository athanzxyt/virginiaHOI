# Pre-Processing Steps

```
## Import Libraries
import pandas as pd
import numpy as np
```

```
## Read in file
filename = 'Health_Opportunity_Index.csv'
df = pd.read_csv(filename)
```

Now we turn the health opportunity index column into 3 sorted categories: good, medium, and bad.

```
## Convert Continuous Label to Categorical Label (Discretiziation)

label_name = 'Health Opportunity Index'
bins = [-0.1,np.percentile(df[label_name],25),np.percentile(df[label_name],75),1]
df[label_name] = pd.cut(df[label_name], bins=bins, labels=['Bad','Medium','Good'])
```

```
## Remove the useless "census track" column
cols = df.columns.tolist()
cols = cols[1:-1]
df = df[cols]
```

```
## One-hot Encoding
df = df.replace(['Urban','Rural'],[1.0,0.0])
```

```
## drop 0 values which shouldn't be there in the 'access to care' column
df.drop(df.index[df['Access to Care'] == 0.0], inplace = True)
```

```
## Export Dataset for Weka Attribute Selection
df.to_csv('HOI_clean.csv',index=False)
```
